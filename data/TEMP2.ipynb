```python
import json
```

```python
import time

current_epoch_seconds = time.time()
```

```python
DATA = {"hard": json.load(open("/Users/akornilo/Downloads/livecodebenchprohard.json"))["llms"],
        "medium": json.load(open("/Users/akornilo/Downloads/livecodebenchmedium.json"))["llms"],
        "easy": json.load(open("/Users/akornilo/Downloads/livecodebencheasy.json"))["llms"]
       }
```

```python
BASE = {}
```

```python
for entry in DATA["hard"]:
    BASE[entry["name"]] = {"hard": entry}
    
for entry in DATA["medium"]:
    BASE[entry["name"]]["medium"] = entry
    
for entry in DATA["easy"]:
    BASE[entry["name"]]["easy"] = entry
```

```python
BASE['gpt-5-2025-08-07']
```

```python
for entry in BASE.values():
    id = entry['hard']['name'].replace("/", "_")
    json.dump({
        "schema_version": "0.0.1",
        "evaluation_id": f"livecodebenchpro/{entry['hard']['name']}/{current_epoch_seconds}",
        "retrieved_timestamp": current_epoch_seconds,
        "source_data_urls": ["https://webhook.cp-bench.orzzh.com/leaderboard/llm/difficulty?difficulty=hard&benchmark_mode=live",
                            "https://webhook.cp-bench.orzzh.com/leaderboard/llm/difficulty?difficulty=medium&benchmark_mode=live",
                            "https://webhook.cp-bench.orzzh.com/leaderboard/llm/difficulty?difficulty=easy&benchmark_mode=live"],
        "leaderboard_name": "Live Code Bench Pro",
        "source_metadata": {
            "source_organization_name": "TBD",
            "evaluator_relationship": "third_party"
        },
        "model_info": {
            "name": entry['hard']['name'],
            "developer": entry['hard']["organization"],
            "inference_platform": entry['hard']["provider"]
        },
        "evaluation_results": [
            {
                "evaluation_name": "Hard Problems" ,
                "metric_config": {
                    "evaluation_description": "Pass@1 on Hard Problems",
                    "lower_is_better": False,
                     "score_type": "continuous",
                     "min_score": 0,
                     "max_score": 1

                },
                "score_details": {
                    "score": entry['hard']["passrate"]
                }
            },
            {
                "evaluation_name": "Medium Problems" ,
                "metric_config": {
                    "evaluation_description": "Pass@1 on Medium Problems",
                    "lower_is_better": False,
                     "score_type": "continuous",
                     "min_score": 0,
                     "max_score": 1

                },
                "score_details": {
                    "score": entry['medium']["passrate"]
                }
            },
            {
                "evaluation_name": "Easy Problems" ,
                "metric_config": {
                    "evaluation_description": "Pass@1 on Easy Problems",
                    "lower_is_better": False,
                     "score_type": "continuous",
                     "min_score": 0,
                     "max_score": 1

                },
                "score_details": {
                    "score": entry['easy']["passrate"]
                }
            }
        ]
    },
        open(f"livecodebenchpro/{id}", 'w'),
        indent=2)


```
